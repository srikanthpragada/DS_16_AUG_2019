```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
```

```python
tips = sns.load_dataset("tips")
```

### Categorical plots

```python
sns.boxplot(x="day",y="tip", data = tips)
```

```python
sns.boxplot(x="day",y="tip", hue="sex", data = tips)
```

```python
sns.barplot(x="day",y="tip", hue="sex", data = tips, ci=False)
```

```python
sns.catplot(x="day",y="total_bill",data=tips, hue="sex",col="smoker", kind='box')
```

```python
sns.catplot(x="day",y="total_bill",data=tips, hue="sex", row="time",col="smoker", kind='box')
```

```python
sns.countplot(x="day",data=tips)
```

```python
flights = sns.load_dataset("flights")
```

```python
flights.groupby('year')['passengers'].mean()
```

```python
sns.barplot(x="year",y="passengers",data=flights,ci=False,estimator=np.sum)
```

### Relational plots

```python
sns.scatterplot(x="total_bill",y="tip", data=tips)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="sex", data=tips)
```

```python
sns.lineplot(x="total_bill",y="tip",hue="sex",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",col="sex",hue="time",data=tips,kind='scatter')
```

```python
sns.lineplot(x="year",y="passengers",hue="month",data=flights, estimator=np.sum)
```

#### pairplot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, vars=['tip','total_bill'])
```

### Regression Plot

```python
sns.regplot(x="total_bill",y='tip', data=tips)
```

```python
sns.lmplot(x="total_bill",y='tip',hue="smoker", col="sex", data=tips)
```

```python
sns.regplot(x="total_bill",y='size', data=tips)
```

```python
sns.residplot(x="total_bill",y='tip', data=tips)
```
